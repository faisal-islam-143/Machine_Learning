Features Scaling And Encoding And Train_Test_Splite:


Look at this list and think in categories:

Pure identifiers:

 Pure identifiers:
  — mostly right, one correction. You said "the model cannot catch any things they can but here are not the case" — a bit tangled, but here's the clean version: yes, all the data in these columns ties to exactly one specific person (an ID, an email), and the key point isn't that the model "can't" use them technically — it can technically process any number or category you feed it — it's that there's no real-world reason these should predict anything. An email address doesn't cause someone to quit their job; it's arbitrary, unrelated to behavior. That's why they're dropped — not a technical limitation, but the absence of a meaningful relationship to the target.

Dates/timestamps:
 — mostly right. You correctly said dates come as text/strings, and models need numbers. One thing to add: it's not just "convert to numeric" in a generic sense — you specifically convert it into something meaningful, like "how many years ago did this happen." A raw date turned into a random number wouldn't help; a date turned into "years of tenure" gives the model something with real predictive value.

Leakage risks:
 — exactly right. Good, clean, accurate as stated. Tax/NetSalary are just Salary renamed/recalculated — same underlying information, so keeping them would let the model cheat.

Genuine numeric features:
 — this one needs correction. You said "columns on which we take the decision" — that's not quite it. The real definition: these are columns where the actual numeric value and its scale matter — like Age (30 vs 60 is a meaningful, orderable difference) or Salary (50,000 vs 500,000 means something specific mathematically). They get scaled specifically because different numeric columns can have wildly different ranges (age 18-65 vs salary 25,000-500,000), and without adjusting for that, columns with naturally bigger numbers could unfairly dominate a model's math, even if they're not actually more important.

"""
==========================================================
REUSABLE DATA CLEANING CHECKLIST
==========================================================
Keep this as a reference for every future dataset.
Two playbooks: NUMERIC columns and CATEGORICAL columns,
plus the one extra rule that applies once you've already
done a train/test split.
==========================================================
"""

# ==========================================================
# NUMERIC COLUMN CHECKLIST
# ==========================================================
"""
1. Look at the shape first — .describe()
   Check min, max, mean vs. median.

2. If stored as text (shows count/unique/top/freq instead
   of mean/min/max) — convert it:
       col = pd.to_numeric(col, errors='coerce')
   Anything unparseable becomes NaN automatically.

3. Spot impossible values using DOMAIN KNOWLEDGE —
   not just statistics. Examples: negative age, salary
   over a sane cap, coordinates outside real range, a
   score above its stated max scale (e.g. 15 on a 1-10 scale).

4. Run the IQR check anyway, as a ROUTINE habit — even if
   nothing looks obviously wrong on the surface:

       Q1 = col.quantile(0.25)
       Q3 = col.quantile(0.75)
       IQR = Q3 - Q1
       lower_bound = Q1 - 1.5 * IQR
       upper_bound = Q3 + 1.5 * IQR

   IMPORTANT LESSON: IQR catches STATISTICAL extremity, not
   LOGICAL impossibility. It missed PerformanceScore's "15"
   on a 1-10 scale because 15 wasn't statistically rare
   enough to fall outside the IQR bounds. Domain knowledge
   must supplement IQR, never replace it. Also: IQR sometimes
   finds nothing at all (e.g. ExperienceYears) — that's a
   valid, legitimate outcome, not a failed check.

5. Replace impossible/outlier values with NaN:
       col = col.replace([bad_val1, bad_val2], np.nan)

6. Recheck .describe() — confirm mean/median now look sane,
   and see the REAL missing percentage (often higher than
   the original count, since fake values just became new NaNs).

7. Decide: impute or drop.
   - Under ~50% missing, sensible distribution -> impute.
   - Choose MEAN if mean ~= median (no real skew).
   - Choose MEDIAN if there's a real gap between them
     (skew present — median is robust to the outliers,
     mean is NOT, since mean is pulled by extreme values).
   - Over ~50% missing -> drop the column entirely.
   - NOTE: imputing a large % of a column with one repeated
     value (mean/median) creates an artificial spike/cluster
     in the data — fine for light missingness, a real
     distortion at ~25%+ missingness. Worth knowing, not
     necessarily worth avoiding for this exercise.

8. TARGET column exception: if this is the column you're
   trying to PREDICT (e.g. Attrition) — NEVER impute it.
   A guessed label corrupts what the model learns to
   predict. Drop rows with a missing target instead:
       df = df.dropna(subset=['target_col'])
"""


# ==========================================================
# CATEGORICAL COLUMN CHECKLIST
# ==========================================================
"""
1. Look at the actual values first:
       col.unique()
       col.value_counts(normalize=True)

2. Check for DISGUISED PLACEHOLDERS — values that look like
   real categories but actually mean "no answer" (e.g. '-',
   'TBD', 'unknown'). Replace these with NaN FIRST, before
   anything else:
       col = col.replace(['-', 'TBD', 'unknown'], np.nan)

3. Check for CASE inconsistency — 'Male'/'male'/'MALE'.
   Fix with .str.title() if it's purely a capitalization issue:
       col = col.str.title()

4. Check for ABBREVIATIONS/SYNONYMS that .str.title() CANNOT
   fix — 'm'/'M' for Male, 'Pak.' vs 'Pakistani', '1'/'0'/
   'Yes'/'No'. These need explicit mapping:
       col = col.replace(['m', 'M'], 'Male')
       col = col.replace(['Pak.'], 'Pakistani')

5. Check for FORMATTING inconsistency — 'Full Time' vs
   'Full-Time'. Pick one style, standardize the whole column:
       col = col.replace(['Full Time'], 'Full-Time')

6. DO NOT erase legitimate minority categories. Something
   like 'Other' is a real answer, not junk, just because it's
   rare. Only remove values that are genuinely fake/meaningless
   (see step 2). Rarity alone is never a reason to delete a
   category.

7. RECHECK .value_counts(normalize=True) AFTER standardizing,
   BEFORE deciding on imputation. Merging variants can change
   which category is actually the true majority (e.g. 'Indeed'
   jumped from ~19% to ~35% once 'INDEED' was merged in) — if
   you impute before merging, you may fill with the wrong value.

8. Decide: impute or drop — same ~50% missingness threshold
   logic as numeric columns.

9. If imputing, use MODE (the categorical equivalent of
   mean/median):
       col = col.fillna(col.mode()[0])
   NOTE: .mode() returns a Series, not a plain value — [0]
   extracts the actual value pandas can use in fillna().

10. TARGET column exception: same as numeric — if this
    column is what you're predicting, NEVER impute it.
    Standardize its categories, then drop rows with missing
    target values instead.
"""


# ==========================================================
# THE ONE EXTRA RULE: WORKING WITH TRAIN/TEST SPLITS
# ==========================================================
"""
Everything above stays the same. The ONLY difference: once
you've split into X_train / X_test, you must apply fixes to
BOTH — but only ever CALCULATE a statistic (mean, median,
mode) from X_train. Never calculate a fresh statistic from
X_test — that leaks test-set information into how you're
preparing your data, and inflates your model's apparent
performance dishonestly.

Pattern to follow every time:

    # 1. Fix fake/bad values — do this in BOTH the same way
    X_train['col'] = X_train['col'].replace([bad_value], np.nan)
    X_test['col']  = X_test['col'].replace([bad_value], np.nan)

    # 2. Calculate the fill value — ONLY from X_train
    fill_value = X_train['col'].mean()      # or .median() or .mode()[0]

    # 3. Apply that SAME number to both
    X_train['col'] = X_train['col'].fillna(fill_value)
    X_test['col']  = X_test['col'].fillna(fill_value)

NEVER write X_test['col'].mean() / .median() / .mode() on its
own — you calculate the statistic exactly once, from training
data, and reuse that same number for the test set too.
"""
