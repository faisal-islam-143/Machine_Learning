Hidden knowledge 
 this is an important connection to make: univariate and bivariate analysis aren't separate stories, they connect. A skewed x-axis variable will always make a scatter plot look bunched up on one side — that's not a flaw in the relationship, it's the skew from Step 1 showing up again in a new plot. This is exactly why real analysts often log-transform skewed variables before scatter-plotting them — it spreads the bunched-up cluster out and makes the true relationship easier to see visually.
